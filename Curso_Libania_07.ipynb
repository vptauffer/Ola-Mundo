```python
import pandas as pd
dados = pd.read_excel('http://profalibania.com.br/ds/secao05video02.xlsx')

print(dados)

dados2 = dados.drop_duplicates() # neste caso só é jogado fora dados duplicados em todas as colunas
print(dados2)

dados3 = dados.sort_values(['Prontuário',"Data admissão"], ascending=False)
# vai listar os dados em ordem decrescente (o mais recente primeiro), para podermos limpar as duplicatas a prtir dele (mais antigos)
dados3 = dados3.drop_duplicates(subset=['Prontuário','Documento'])
print(dados3)

# ainda assim, podemos continuar com documentos duplicados "escondidos". Então pedimos para o pandas informar os duplicados
dados3['Duplicado'] = dados3.duplicated(['Documento'])
print(dados3)  # mostra que existem 2 prontuários com o mesmo documento

dados3['Duplicado'] = dados3.duplicated(['Documento'],keep=False)  # keep=False vai evidenciar os dois documentos iguais
print(dados3)

# Se eu quiser filtrar somente duplicados:
paraRevisao = dados3[dados3['Duplicado']==True]
print (paraRevisao)
```

```python
# Limpando os vazios (NaN)- importante sempre checar com a organização antes de deletar dados faltantes (se vale a pena ir atrás daquele dado)
dados4 = dados3.dropna()
print(dados4)

# ou podemos substituir dados faltantes por outra coisa:
dados5 = dados3.fillna(-1)
print(dados5)

# também podemos preencher seletivamente:
dados5 = dados3.fillna({"CEP":-1,"Fonte pagadora":"sem resposta"}) # chaves {} indica que você está criando uma espécie de dicionário
print(dados5)

# o pandas lida com números como float (.0), causando problema no caso do CEP
dados5['CEP'] = dados5['CEP'].astype(int)  # transforma para número inteiro
dados5['CEP'] = dados5['CEP'].astype(str)  # transforma em string (caracteres)

dados5.loc[dados5['CEP'] < '0', 'CEP'] = 'não informado' # localiza os -1 e substitui por não informado
dados5

```
