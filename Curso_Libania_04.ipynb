```python
# Colunas no Pandas


import pandas as pd
dados = pd.read_excel('http://profalibania.com.br/ds/secao04video03.xlsx')
dados[['Data','Dia da semana']]  # manipula algumas colunas sem precisar chamar todas

# Quero criar uma coluna com o ano de nascimento dos pacientes
dados['Ano Admissao'] = dados['Data'].dt.year
dados['Ano Nasc'] = dados ['Ano Admissao'] - dados['Idade']
dados
# podemos fazer tudo em 1 linha:
dados['Ano Nasc 2'] = dados['Data'].dt.year - dados['Idade']
dados

# podemos apagar uma coluna que não desejamos mais:
dados = dados.drop(['Ano Admissao'], axis=1)  # por padrão a deleção é de linha. axis=1 significa coluna
dados
```
